```python
import matplotlib.pyplot as plt
import numpy as np
```

Loading data

```python
ecgdata = r'C:\Users\Helies\Desktop\M2\S9\Sant√© AP\CM1_MOTTET_Denis\ECG_Analysis\data\ECG_Hz.txt'
data = np.loadtxt(ecgdata)
```

Parameters and pre-process

```python
data0 = data[:,0] # Extraction of the Lead I data
data1 = data[:,1] # Extraction of the Lead II data
data2 = data[:,2] # Extraction of the Lead III data

data0_min = min(data0) # Define the minima of the data0 matrix
data1_min = min(data1) # Define the minima of the data1 matrix
data2_min = min(data2) # Define the minima of the data2 matrix

data0_max = max(data0) # Define the maxima of the data0 matrix
data1_max = max(data1) # Define the maxima of the data1 matrix
data2_max = max(data2) # Define the maxima of the data2 matrix

data0_norm = (data0-data0_min)/(data0_max-data0_min) # Normalize data0
data1_norm = (data1-data1_min)/(data1_max-data1_min) # Normalize data1
data2_norm = (data2-data2_min)/(data2_max-data2_min) # Normalize data2

sampling = 1024 # Based on the given informations sampling rate is 1024 Hz
time = np.arange(0, len(data0)/sampling, 1/sampling) # Time vectore based on sampling parameter
```

Ploting

```python
plt.subplot(3,1,1)
plt.plot(time, data0_norm) # Ploting the data from the time vectore and the Lead I
plt.axhline(y=1, color='r', linestyle='-') # Plot a red horizontal line at the maxima
plt.axhline(y=0.5, color='r', linestyle='-') # Plot a red horizontal line at 0.5
plt.axhline(y=0, color='r', linestyle='-') # Plot a red horizontal line at the minima

plt.xlabel('Time (s)') # Labeling the x axis as a time vector
plt.ylabel('Lead I (mV)') # Labeling the y axis as a mV vector
plt.grid(color ='grey', linestyle='--') # Plot the grid


plt.subplot(3,1,2)
plt.plot(time, data1_norm) # Ploting the data from the time vectore and the Lead II
plt.axhline(y=1, color='r', linestyle='-') # Plot a red horizontal line at the maxima
plt.axhline(y=0.5, color='r', linestyle='-') # Plot a red horizontal line at 0.5
plt.axhline(y=0, color='r', linestyle='-') # Plot a red horizontal line at the minima

plt.xlabel('Time (s)') # Labeling the x axis as a time vector
plt.ylabel('Lead II (mV)') # Labeling the y axis as a mV vector
plt.grid(color ='grey', linestyle='--') # Plot the grid


plt.subplot(3,1,3)
plt.plot(time, data2_norm) # Ploting the data from the time vectore and the Lead III
plt.axhline(y=1, color='r', linestyle='-') # Plot a red horizontal line at the maxima
plt.axhline(y=0.5, color='r', linestyle='-') # Plot a red horizontal line at 0.5
plt.axhline(y=0, color='r', linestyle='-') # Plot a red horizontal line at the minima

plt.xlabel('Time (s)') # Labeling the x axis as a time vector
plt.ylabel('Lead III (mV)') # Labeling the y axis as a mV vector
plt.grid(color ='grey', linestyle='--') # Plot the grid 
```
